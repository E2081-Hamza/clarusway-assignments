```python
def factor(n,top = 1):
    """ calculates factorial of given"""
    for i in range(1,n+1):
        top = top * i
    return top
def flwless_entry():
    """to avoid invalid entry for factor function"""
    try:
        a = int(input("Enter an integer for output:"))
        if a >= 0 :
            return factor(a)
        else:
            print("pls enter an positive integer")
    except ValueError:
        print("pls enter an positive integer")
print(flwless_entry())
```
